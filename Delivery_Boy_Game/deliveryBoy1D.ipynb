# Delivery Boy #

This game is about a <b>delivery boy</b>, riding a bike, who has got a job to <b>deliver</b> a <b>parcel</b> to it's intended <b>receiver</b>. This is a 1-dimensional game where the length of the world has been set to 50 and the delivery boy has to deliver the parcel in atmost 50 steps. On his way, he will encounter various <b>percepts</b>(the percepts have been located randomly in the world) like <b>Gas Station</b>, <b>Food Court</b>, <b>Helmet</b>, etc. Based on the given percept, the agent will perform a respective action. For e.g. If the delivery boy <b>percepts</b> <b>Food</b> court, he will <b>eat</b> food. Also, if the agent(delivery boy) refuels his bike, he will get 25 steps extra to ride his bike in the world. This increases the chance for the agent(delivery boy) to deliver the parcel.

Defined required global variables.

```python
# Global Variables
movesTaken = 0# to count the number of moves the agent takes to deliver the parcel.
ateFood = 0# to check if the agent(delivery boy) ate food.
doneDelivery = 0# to check if the agent(delivery boy) has done the delivery.
bikeRefueled = 0# to check if the agent(delivery boy) has refueled his bike.
```

In the below cell, defined <b>Delivery_Boy</b> Class subclassing Agent class from agents.py module.
The instances of this class are the actions that the agent will perform as he encounters a percept as per the table below:

```python
class Delivery_Boy(Agent):
    location = 1# Initializing the class variable location.
    
    def move_down(self):
        self.location += 1# Incrementing the location varibale by 1
        
    def eat(self, thing):# This function is called when the agent(delivery boy) percepts food at it's currect location.
        
        if isinstance(thing, Food):# if the agent(delivery boy) finds food
            print("Delivery Boy: Ate food at {}.".format(self.location))
            # printing Agent(delivery boy) ate food at currect location.
            return True
        return False
    
    def refuel(self, thing):# This function is called when the agent(delivery boy) percepts a Gas station at it's current location.
        
        if isinstance(thing, Gas):# if the agent(delivery boy) finds Gas station.
            print("Delivery boy: Refueled the bike at {}.".format(self.location))
            # printing agent(delivery boy) refueled his bike at it's currect location.
            return True
        return False
        
    def slow_down(self, thing):# This function is called where the agent(delivery boy) percepts a Pit at it's currect location.
        
        if isinstance(thing, Pit):# if the agent(delivery boy) finds a pit.
            print("Delivery boy: Slowed down bike at {}.".format(self.location))
            # printing agent(delivery boy) slowed down his bike to cross the pit.
            return True
        return False
    
    def deliver_parcel(self, thing):# This function is called where the agent(delivery boy) percepts a Destination at it's current location.
        
        if isinstance(thing, Destination):# if the agent(delivery boy) finds his destincation.
            print("Delivery boy: Delivered parcel at {}.".format(self.location))
            # printing agent(delivery boy) delivered the parcel at it's destination.
            return True
        return False
```

Now, defined <b>Road</b> class which is taking <b>Environment</b> as subclass which will control the actions that the agent will perform based on what he percepts from the environment. It also contains an instance <b>is_done</b> which will check/ensure the conditions under which the game is complete. Those conditions are as below:
1. There is no destination left  unvisited in the world i.e. the agent has delivered the parcel at it's destination.
2. The agent is dead.

```python
class Road(Environment):
    def percept(self, agent):
    # This fucntion returns the list of percepts that the agent(delivery boy) finds at it's currect location.
        
        things = self.list_things_at(agent.location)
        # getting the things(objects/percepts) from agent's currect location.
        return things
    
    def execute_action(self, agent, action):
    # This function is called when the agent needs to execute an action based on the currect percept.
        
        global movesTaken
        global ateFood
        global doneDelivery
        global bikeRefueled
        
        movesTaken += 1# Incrementing the movesTaken variable by 1
        if action == "move_down":# if the agent(delivery boy) decides to move down.
            
            print('{} decided to {} at location: {}'.format(str(agent)[1:-1], action, agent.location))
            agent.move_down()# Calling the move_down function.
            
        elif action == "eat":# if the agent(delivery boy) decides to have food.
            
            items = self.list_things_at(agent.location, tclass=Food)# returns the items(object) present at the agent's currect location.
            
            if len(items) != 0:
                if agent.eat(items[0]):# if the agent(delivery boy) eats item(food)
                    
                    ateFood = 1
                    print('{} ate {} at location: {}'.format(str(agent)[1:-1], str(items[0])[1:-1], agent.location))
                    self.delete_thing(items[0])# deleting the item(food) from the world.
                    
        elif action == "refuel":# if the agent(delivery boy) decides to refuel his bike.
            items = self.list_things_at(agent.location, tclass=Gas)# returns the thing(object) present at the agent's currect location.
            if len(items) != 0:
                if agent.refuel(items[0]):# if the agent(delivery boy) refuels his bike.
                    
                    bikeRefueled = 1
                    print('{} refueled the bike {} at location: {}'.format(str(agent)[1:-1], str(items[0])[1:-1], agent.location))
                    self.delete_thing(items[0])# deleting the item(gas) from the world.
                    
        elif action == "slow_down":# if the agent(delivery boy) decides to slow down while crossing the pit.
            
            items = self.list_things_at(agent.location, tclass=Pit)# returns the thing(object) present at the agent's currect location.
            
            if len(items) != 0:
                if agent.slow_down(items[0]):# if the agent(delivery boy) slows down his bike.
            
                    print('{} found {} at location: {}'.format(str(agent)[1:-1], str(items[0])[1:-1], agent.location))
                    self.delete_thing(items[0])# deleting the item(pit) from the world.
                    
        elif action == "deliver_parcel":# if the agent decides to deliver parcel at it's destination.
            
            items = self.list_things_at(agent.location, tclass=Destination)
            
            if len(items) != 0:
                if agent.deliver_parcel(items[0]):# if the agent delivers parcel at destination.
                    
                    doneDelivery = 1
                    print('{} {} at location: {}'.format(str(agent)[1:-1], str(items[0])[1:-1], agent.location))
                    self.delete_thing(items[0])# deleting the item(destination) from the world.  
                    
    def is_done(self):
    # This function is called to check if the agent is alive or the destination location has been visited.
        
        no_destinations = not any(isinstance(thing, Destination) for thing in self.things)
        # returns true if the destination location has been visited
        
        dead_agents = not any(agent.is_alive() for agent in self.agents)
        # returns true if the agent is not alive anymore.
        
        return dead_agents or no_destinations
```

This instance will return the desired action to be performed by the agent based on the given percept.

```python
def program(percepts):
# This function returns a function based on the given percept.
    
    for p in percepts:
        if isinstance(p, Food):# if the agent percepts food.
            return 'eat'# returns action : eat
        
        elif isinstance(p, Gas):# if the agent percepts Gas station.
            return 'refuel'# returns action : refuel
        
        elif isinstance(p, Destination):# if the agent percepts Destination.
            return 'deliver_parcel'# returns action : deliver_parcel
        
        elif isinstance(p, Pit):# if the agent percepts a pit.
            return 'slow_down'# returns action : slow_down
        
    return 'move_down'# if the agent(delivery boy) percepts nothing, return action : move_down
```

This function is responsible for the creation of objects of the percept classes and definig location of agent and various percepts in the world.

```python
def runTheGame(Environment):
    
    steps = 25# Number to steps/moves an agent can take in one run.
    
    dBoy = Delivery_Boy(program)# Creating object of the class Delivery_Boy.
    food = Food()# Creating obejct of the class Food.
    pit = Pit()# Creating object of the class Pit.
    gas = Gas()# Creating object of the class Gas.
    destination = Destination()# Creating object of the class Destination.
    road = Road()# Creating object of the class Road
    
    delBoyStartPoint = 1# specifying the start point of agent(delivery boy).
    
    foodLoc = random.randint(2, 50)# taking random location of food.
    gasLoc = random.randint(2, 50)# taking random location of Gas station.
    pitLoc = random.randint(2, 50)# taking random location of pit.
    destLoc = random.randint(2, 50)# taking random location of the destination.
    
    # Making sure the location of destination does not overlap with the location of any other percept/object.
    while(destLoc == foodLoc 
          or destLoc == gasLoc
          or destLoc == pitLoc):
        foodLoc = random.randint(2, 50)
        gasLoc = random.randint(2, 50)
        pitLoc = random.randint(2, 50)
        destLoc = random.randint(2, 50)
    
    road.add_thing(dBoy, delBoyStartPoint)# Adding agent(delivery boy) at it's starting point.
    road.add_thing(food, foodLoc)# Adding food at it's randomized location.
    road.add_thing(gas, gasLoc)# Adding gas station at it's randomized location.
    road.add_thing(pit, pitLoc)# Adding pit at it's randomized location.
    road.add_thing(destination, destLoc)# Adding destination at it's randomized location.
    
    print("###################################")
    print("Length of World : ", 50)
    print("Location of Agent : ", delBoyStartPoint)
    print("Location of Food : ", foodLoc)
    print("Location of Gas : ", gasLoc)
    print("Location of Pit : ", pitLoc)
    print("Location of Destination : ", destLoc)
    print("###################################")
    
    road.run(steps)
    # make the agent(deliver boy) to run for defined number of steps i.e. 25
    
    if bikeRefueled == 1:# check if the agent has refueled his bike. If yes,
        road.run(steps)# make the agent(deliver boy) to run for defined number of steps i.e. 25
    
    if doneDelivery == 0:# If the agent(delivery boy) couldn't deliver the parcel.
        print("Bike ran out of fuel. Delivery Boy couldn't deliver the parcel.")
        
    print("Total number of moves : ",movesTaken)
    # printing total number of moves taken.
```

Running the game and calculating performance measure.

```python
movesTaken = 0# re-initializing the value of variable movesTaken so that the count starts again from 0 for the next run.
runTheGame(Environment)# calling the runTheGame function.

perf_measure = (200*doneDelivery) - (10*ateFood) - 10*(bikeRefueled) - movesTaken# calculating the performance measure
print("####################################")
print("Performance Measure of Delivery Boy : ", perf_measure, " out of 200")
print("####################################")
# printing performance measure
```

```python
class Delivery_Boy(Agent):
    location = [0,0]# Initializing the class variable location.
    direction = Direction("down")# Initializing the class variable direction indicating where the agent(delivery boy) is facing.
    
    def moveforward(self, success):# This function is called when the agent(delivery boy) decides to move forward.
        
        global turn
        if not success:# if the agent hits the boundary.
            turn = True # for the agent(delivery boy) to remember to turn.
            return
        if self.direction.direction == Direction.R:
            self.location[0] += 1# Incrementing x-axis of location by 1 if direction is rightwards
        elif self.direction.direction == Direction.L:
            self.location[0] -= 1# Decrementing x-axis of location by 1 if direction is leftwards
        elif self.direction.direction == Direction.D:
            self.location[1] += 1# Incrementing y-axis of location by 1 if direction is downwards
        elif self.direction.direction == Direction.U:
            self.location[1] -= 1# Decrementing y-axis of location by 1 if direction is upwards
    
    def turn(self, direct):# This fucntion is called when the agent(delivery boy) needs to turn left or right.
        self.direction = self.direction + direct# changes the currect direction of the agent(delivery boy) to the desired direction.
        # Here, direct is the direction where the agent is turning i.e. either left or right
        
    def eat(self, thing):# This function is called when the agent(delivery boy) percepts food at it's currect location.
        
        if isinstance(thing, Food):# if the agent(delivery boy) finds food
            print("Delivery Boy: Ate food at {}.".format(self.location))# printing Agent(delivery boy) ate food at currect location.
            return True
        return False
    
    def refuel(self, thing):# This function is called when the agent(delivery boy) percepts a Gas station at it's current location.
        
        if isinstance(thing, Gas):# if the agent(delivery boy) finds Gas station.
            refueled = 1
            print("Delivery Boy: refueled bike at {}.".format(self.location))# printing agent(delivery boy) refueled his bike at it's currect location.
            return True
        return False
    
    def slow_down(self, thing):# This function is called where the agent(delivery boy) percepts a Pit at it's currect location.
        
        if isinstance(thing, Pit):# if the agent(delivery boy) finds a pit.
            print("Delivery Boy: slowed down bike at {}.".format(self.location))# printing agent(delivery boy) slowed down his bike to cross the pit.
            return True
        return False
    
    def deliver_parcel(self, thing):# This function is called where the agent(delivery boy) percepts a Destination at it's current location.
        
        if isinstance(thing, Destination):# if the agent(delivery boy) finds his destincation.
            print("Delivery Boy: delivered parcel at {}.".format(self.location))# printing agent(delivery boy) delivered the parcel at it's destination.
            return True
        return False
    
    def wear(self, thing):# This function is called when the agent(delivery boy) percepts a helmet at it's current location.
        
        if isinstance(thing, Helmet):# if the agent(delivery boy) finds a helmet.
            print("Delivery Boy: wore Helmet at {}.".format(self.location))# printing agent(delivery boy) picked up helmet from it's currect location.
            return True
        return False
    
    def go_on(self, thing):# This function is called when the agent(delivery boy) is wearing helmet while he percepts a pit at it's curent location.
        
        if isinstance(thing, Pit):# if the agent(delivery boy) percepts a pit.
            print("Delivery Boy: Rode through the pit using helmet at {} and broke the helmet.".format(self.location))
            # printing agent(delivery boy) rode through the pit(without slowing down) and broke his helmet without himself getting hurt.
            return True
        return False
    
    def call_receiver(self, thing):# This function is called when the receiver is not at his home to collect the parcel.
        if isinstance(thing, Destination):# if the agent(delivery boy) percepts destination.
            print("The receiver is not at his home. Calling the receiver....... from location {}".format(self.location))
            # printing agent(delivery boy) calling the receiver from his current location.
            return True
        return False
```

Road2D is a two-dimensional environment class having "GraphicEnvironment" as subclass from "agents.py" module.
It also contains an instance <b>is_done</b> which will check/ensure the conditions under which the game is complete. Those conditions are as below:
1. There is no destination left unvisited in the world i.e. the agent has delivered the parcel at it's destination.
2. The agent is dead.

```python
class Road2D(GraphicEnvironment):
    
    def percept(self, agent):# This fucntion returns the list of percepts that the agent(delivery boy) finds at it's currect location.
        
        things = self.list_things_at(agent.location)# getting the things(objects/percepts) from agent's currect location.
        return things

    def execute_action(self, agent, action):# This function is called when the agent needs to execute an action based on the currect percept.
        
        global movesTaken
        global woreHelmet
        global ateFood
        global doneDelivery
        global bikeRefueled
        
        movesTaken += 1# Incrementing the moves variable by 1
        
        if action == 'turnright':# if the agent(delivery boy) decides to turn right.
            
            print('{} decided to {} at location: {}'.format(str(agent)[1:-1], action, agent.location))
            agent.turn(Direction.R)
            # Making agent(delivery boy) to turn right.
            
        elif action == 'turnleft':# if the agent(delivery boy) decides to turn left
            
            print('{} decided to {} at location: {}'.format(str(agent)[1:-1], action, agent.location))
            agent.turn(Direction.L)
            # Making agent(delivery boy) to turn left.
            
        elif action == 'moveforward':# if the agent(delivery boy) decides to move forward.
            
            loc = copy.deepcopy(agent.location) # Copying the location of agent(delivery boy) to loc
            
            if agent.direction.direction == Direction.R:# if the agent is moving forward after turning right
                
                loc[0] += 1# Incrementing x-axis of loc by 1
                
            elif agent.direction.direction == Direction.L:# if the agent is moving forward after turning left
                
                loc[0] -= 1# Decrementing x-axis of loc by 1
                
            elif agent.direction.direction == Direction.D:# if the agent is moving forward after turning down
                
                loc[1] += 1# Incrementing y-axis of loc by 1
                
            elif agent.direction.direction == Direction.U:# if the agent is moving forward after turning up
                
                loc[1] -= 1# Decrementing y-axis of loc by 1
                
            if self.is_inbounds(loc):# check if the target location is falling inside our world.
                
                print('{} decided to move {}wards at location: {}'.format(str(agent)[1:-1], agent.direction.direction, agent.location))
                agent.moveforward(True)# calling moveforward function with success = True.
                
            else:
                
                print('{} decided to move {}wards at location: {}, but couldnt'.format(str(agent)[1:-1], agent.direction.direction, agent.location))
                agent.moveforward(False)# calling moveforward function with success = False.
                
        elif action == "eat":# if the agent decides to have food.
            
            items = self.list_things_at(agent.location, tclass=Food)# returns the items(object) present at the agent's currect location.
            
            if len(items) != 0:
                if agent.eat(items[0]):# if the agent eats item(food)
                    
                    ateFood = 1
                    print('{} ate {} at location: {}'.format(str(agent)[1:-1], str(items[0])[1:-1], agent.location))
                    self.delete_thing(items[0])
                    #deleting the item(food) from the world.
                    
        elif action == "refuel":# if the agent decides to refuel his bike.
            
            items = self.list_things_at(agent.location, tclass=Gas)# returns the thing(object) present at the agent's currect location.
            
            if len(items) != 0:
                if agent.refuel(items[0]):# if the agent refuels his bike.
                    
                    bikeRefueled = 1
                    print('{} refueled {} at location: {}'.format(str(agent)[1:-1], str(items[0])[1:-1], agent.location))
                    self.delete_thing(items[0])
                    #deleting the item(gas) from the world.
                    
        elif action == "deliver_parcel":# if the agent decides to deliver parcel at it's destination.
            
            items = self.list_things_at(agent.location, tclass=Destination)
            
            if len(items) != 0:
                if agent.deliver_parcel(items[0]):# if the agent delivers parcel at destination.
                    
                    doneDelivery = 1
                    print('{} delievred parcel at {} at location: {}'.format(str(agent)[1:-1], str(items[0])[1:-1], agent.location))
                    self.delete_thing(items[0])
                    #deleting the item(destination) from the world.
                    
        elif action == "slow_down":# if the agent decides to slow down while crossing the pit.
            
            items = self.list_things_at(agent.location, tclass=Pit)# returns the thing(object) present at the agent's currect location.
            
            if len(items) != 0:
                if agent.slow_down(items[0]):# if the agent slows down it's bike.
                    
                    print('{} found {} at location: {}'.format(str(agent)[1:-1], str(items[0])[1:-1], agent.location))
                    self.delete_thing(items[0])
                    #deleting the item(pit) from the world.
                    
        elif action == "wear":# if the agent decides to wear the helmet.
            
            items = self.list_things_at(agent.location, tclass=Helmet)# returns the thing(object) present at the agent's currect location.
            
            if len(items) != 0:
                if agent.wear(items[0]):# if the agent wears a helmet.
                    
                    woreHelmet = 1
                    print('{} picked up {} at location: {}'.format(str(agent)[1:-1], str(items[0])[1:-1], agent.location))
                    self.delete_thing(items[0])
                    #deleting the item(helmet) from the world.
                    
        elif action == "go_on":# if the agent decides not to slow down while crossing the pit.
            
            items = self.list_things_at(agent.location, tclass=Pit)# returns the thing(object) present at the agent's currect location.
            
            if len(items) != 0:
                if agent.go_on(items[0]):# if the agent crosses pit without slowing down.
            
                    print('{} used helmet while riding through {} at location: {}'.format(str(agent)[1:-1], str(items[0])[1:-1], agent.location))
                    woreHelmet = 0# When the agent crosses pit without slowing down(helmet breaks)
                    self.delete_thing(items[0])
                    #deleting the item(pit) from the world.
                    
        elif action == "call_receiver":# if the agent decides to make a call to the receiver.
            
            items = self.list_things_at(agent.location, tclass=Destination)# returns the thing(object) present at the agent's currect location.
            
            if len(items) != 0:
                if agent.call_receiver(items[0]):# if the agent called the receiver
                    
                    global receiverAtHome
 
                    destination = Destination()# creating object of class Destination.
                    destLoc = [random.randint(0,dimension-1), random.randint(0, dimension-1)]# taking random location of the new destination.
                    print('Receiver told {} to deliver the parcel at location: {}'.format(str(agent)[1:-1], destLoc))
                    
                    receiverAtHome = 1# Confirming that the receiver is present at his new location to receive the parcel.
                    self.delete_thing(items[0])# deleting the current destination from the world.
                    self.add_thing(destination, destLoc)# Adding new destination at it's randomized location.
                    
    def is_done(self):
    # This function is called to check if the agent is alive or the destination location has been visited.
        no_destinations = not any(isinstance(thing, Destination) for thing in self.things)
        # returns true if the destination location has been visited
        
        dead_agents = not any(agent.is_alive() for agent in self.agents)
        # returns true if the agent is not alive anymore.
        
        return no_destinations or dead_agents
```

This is the program function which tells what action to take based on the percept.

```python
def program(percepts):
# This function returns a function based on the given percept.
    global turn
    for p in percepts: 
        if isinstance(p, Pit):# if the agent percepts a pit.
            if woreHelmet == 1:# if the agent is wearing a helmet
                return 'go_on'# returns action : go_on(without slowing down)
            else:
                return 'slow_down'# returns action : slow_down
        
        elif isinstance(p, Food):# if the agent percepts food.
            return 'eat'# returns action : eat
        
        elif isinstance(p, Destination):# if the agent percepts Destination.
            if receiverAtHome == 1:
                return "deliver_parcel"# returns action : deliver_parcel
            else:
                return "call_receiver"# returns action : call_receiver
            
        elif isinstance(p, Gas):# if the agent percepts Gas station.
            return "refuel"# returns action : refuel
        
        elif isinstance(p, Helmet):# if the agent percepts helmet.
            return "wear"# returns action : wear
        
    if turn: # if the agent is at the edge of world and has to turn
        turn = False
        choice = random.choice((1,2));# 1- right, 2-left (50% probability each)
    else:
        choice = random.choice((1,2,3,4)) # 1-right (25% probability), 2-left(25% probability), others-forward(50% probability)
    
    if choice == 1:
        return 'turnright'# returns action : turnright
    
    elif choice == 2:
        return 'turnleft'# returns action : turnleft
    
    else:
        return 'moveforward'# returns action : moveforward
```

This is "runTheGame" function having subclass as "GraphicEnvironment" in order to call it's instances for placing the various objects in the world and make the agent move from one location to another.

```python
def runTheGame(GraphicEnvironment):
    
    steps = 250# Number to steps/moves an agent can take in one run.
    
    dBoy = Delivery_Boy(program)# Creating object of the class Delivery_Boy.
    food = Food()# Creating obejct of the class Food.
    pit = Pit()# Creating object of the class Pit.
    gas = Gas()# Creating object of the class Gas.
    helmet = Helmet()# Creating object of the class Helmet.
    destination = Destination()# Creating object of the class Destination.
    
    global dimension# Referring to the global variable 'dimension'.
    
    road = Road2D(dimension, dimension , color={'Delivery_Boy': (52,152,219), # Blue
                                                'Food': (40,180,99), # Green
                                                'Pit': (23,32,42), # Black
                                                'Helmet': (244,208,63), # Yellow
                                                'Destination':(187,143,206), # Purple 
                                                'Gas':(230,126,34) # Orange
                                               })
    # Creating object of the class Road2D and initiating GraphicEnvironment class variables
    # such as width, height and colors of the percepts/objects in the worls.
 
    delBoyStartPoint  = [random.randint(0,dimension-1), random.randint(0, dimension-1)]
    # taking random location of the agent(delivery boy)
    
    foodLoc = [random.randint(0,dimension-1), random.randint(0, dimension-1)]
    # taking random location of food.
    
    gasLoc = [random.randint(0,dimension-1), random.randint(0, dimension-1)]
    # taking random location of Gas station.
    
    pitLoc = [random.randint(0,dimension-1), random.randint(0, dimension-1)]
    # taking random location of pit.
    
    helmetLoc = [random.randint(0,dimension-1), random.randint(0, dimension-1)]
    # taking random location of helmet.
    
    global destLoc 
    destLoc = [random.randint(0,dimension-1), random.randint(0, dimension-1)]
    # taking random location of the destination.
    
    # Making sure the location of destination does not overlap with any other percept/object.
    # Also, making sure, the location of pit and helmet is different.
    while(destLoc == foodLoc 
          or destLoc == gasLoc
          or destLoc == pitLoc
          or destLoc == delBoyStartPoint
          or destLoc == helmetLoc
          or pitLoc == helmetLoc):
        foodLoc = [random.randint(0,dimension-1), random.randint(0, dimension-1)]
        gasLoc = [random.randint(0,dimension-1), random.randint(0, dimension-1)]
        pitLoc = [random.randint(0,dimension-1), random.randint(0, dimension-1)]
        destLoc = [random.randint(0,dimension-1), random.randint(0, dimension-1)]
        helmetLoc = [random.randint(0,dimension-1), random.randint(0, dimension-1)]
    
    road.add_thing(dBoy, delBoyStartPoint)# Adding agent(delivery boy) at it's randomized location.
    road.add_thing(food, foodLoc)# Adding food at it's randomized location.
    road.add_thing(gas, gasLoc)# Adding gas station at it's randomized location.
    road.add_thing(pit, pitLoc)# Adding pit at it's randomized location.
    road.add_thing(helmet, helmetLoc)# Adding helmet at it's randomized location.
    road.add_thing(destination, destLoc)# Adding destination at it's randomized location.
    
    print("###################################")
    print("Dimension of World : ", [dimension,dimension])
    print("Location of Agent : ", delBoyStartPoint)
    print("Location of Food : ", foodLoc)
    print("Location of Gas : ", gasLoc)
    print("Location of Pit : ", pitLoc)
    print("Location of Helemt : ", helmetLoc)
    print("Location of Destination : ", destLoc)
    print("###################################")
    
    road.run(steps)
    # make the agent(deliver boy) to run for defined number of steps i.e. 250
    
    if bikeRefueled == 1:# check if the agent has refueled his bike. If yes,
        road.run(steps)# make the agent(deliver boy) to run for defined number of steps i.e. 250
    
    if doneDelivery == 0:# if the agent(delivery boy) couldn't deliver the parcel
        print("Bike ran out of fuel. Delivery Boy couldn't deliver the parcel.")
    
    print("Total number of moves : ",movesTaken)
    # printing total number of moves taken.
```
